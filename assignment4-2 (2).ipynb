## Part 1:  SVM with no bias term 

Formulate a soft-margin SVM without the bias term, i.e. one where the discriminant function is equal to $\mathbf{w}^{T} \mathbf{x}$.
Derive the saddle point conditions, KKT conditions and the dual.
Compare it to the standard SVM formulation that was derived in class.


## Derivation

For a soft-margin SVM without the bias term, the discriminant function is given as:
$$f(x)=w^Tx$$
The margin of a linear discriminant: $\frac{1}{2}\hat{w}^T(x_{\oplus}-x_{\ominus})$ 
where, $\hat{w}$ is a unit vector in the direction of w

Suppose that $x_{\oplus}$ and $x_{\ominus}$ are equidistant from the descision boundary.
$$w^Tx_{\oplus}=a$$ and $$w^Tx_{\ominus}=-a$$
Subtracting the two equations, we get:
$$w^T(x_{\oplus}-x_{\ominus})=2a$$
Now, divide by the norm of w:
$$\hat{w}^T(x_{\oplus}-x_{\ominus})=\frac{2a}{||w||}$$

Set a=1, we get: $$Margin=\frac{1}{||w||}$$

We know that,
Maximizing the margin is equal to minimizing $$||w||^2$$
Thus,
$$ \min_{w}   \frac{1}{2}  ||w||^2 $$ 
subject to: $$y_i(w^Tx_i)\geq 1,   i=1,2,...N  $$

For soft-margin SVM, we need to include slack variable. $\sum_{i=1}^{N}\xi_i$ is bound on the number of misclassified examples.
Therefore, Optimization problem becomes, 
$$\min \frac{1}{2}||w||^2 + C\sum_{i=1}^{N}\xi_i$$ 
subject to: $ y_i(w^Tx_i)\geq1-\xi_i$; $\xi_i\geq0$; for $i=1,2,...,N$
where C is the penalty parameter.

The **Lagrangian** can be expressed as:
$$ \Lambda (w,\xi,\alpha,\beta)= \frac{1}{2} ||w||^2 + C \sum_{i=1}^{N} \xi_i + \sum_{i=1}^{N} \alpha_i(1-\xi_i-y_iw^Tx_i) -\sum_{i=1}^{N}\xi_i\beta_i$$
where $\alpha_i$ and $\beta_i$ are Lagrange multiplers

**Saddle Point Conditions:**
For soft-margin SVM, saddle point conditions are as follows:
$$ \frac{\partial\Lambda}{\partial w}= w+ \sum_{i=1}^{N}\alpha_i(-y_ix_i)=0$$
Therefore,    $$ w=\sum_{i=1}^{N}\alpha_i y_i x_i$$
$$ \frac{\partial\Lambda}{\partial \xi_i}=C-\alpha_i-\beta_i=0$$

**KKT Conditions:**
For soft-margin SVM, KKT conditions are as follows:


$\nabla\Lambda(x,\lambda)=0$ ... 1st order of derivative of optimality

$g(x)\leq0$ ... Prime constraints

$\lambda\geq0$  ... Dual Constraints

$\lambda g(x)=0$ ... Complimentary slackness condition

$$\frac{\partial\Lambda}{\partial w}=0$$
$$1 - \xi_i - y_i(w^Tx_i)\leq0$$
$$ \alpha_i [1 - \xi_i - y_i(w^Tx_i)] = 0 $$
Since, $ \alpha_i > 0 $
$$ y_i(w^Tx_i) = 1 - \xi_i $$

**Dual Equation:**
We have Lagrangian equation as:
$$ \Lambda (w,\xi,\alpha,\beta)= \frac{1}{2} ||w||^2 + C \sum_{i=1}^{N} \xi_i + \sum_{i=1}^{N} \alpha_i(1-\xi_i-y_iw^Tx_i) -\sum_{i=1}^{N}\xi_i\beta_i$$
using $w=\sum_{i=1}^{N}\alpha_iy_ix_i$
$$w(\alpha)= \frac{1}{2}(\sum_{i=1}^{N} \alpha_iy_ix_i)^T \sum_{j=1}^{N} \alpha_jy_jx_j + C\sum_{i=1}^{N}\xi_i + \sum_{i=1}^{N}\alpha_i- \sum_{i=1}^{N}\xi_i\alpha_i- \sum_{i=1}^{N}\alpha_iy_iw^Tx_i -  \sum_{i=1}^{N}\xi_i\beta_i$$
Therefore, $w(\alpha)= \sum_{i=1}^{N}\alpha_i+\frac{1}{2}(\sum_{i=1}^{N} \alpha_iy_ix_i)^T \sum_{j=1}^{N} \alpha_jy_jx_j -\sum_{i=1}^{N}\alpha_iy_iw^Tx_i + (C-\alpha_i-\beta_i)\sum_{i=1}^{N}\xi_i$

Saddle point condition is, $C-\alpha_i-\beta_i=0$

So, Dual is:
$$\max_{\alpha} \sum_{i=1}^{N}\alpha_i - \frac{1}{2}\sum_{i=1}^{N}\sum_{j=1}^{N}\alpha_i\alpha_jy_iy_jx_i^Tx_j$$
subject to $ \alpha_i\geq0,$ $\beta_i\geq0,$ $\sum_{i=1}^{N}\alpha_iy_i=0,$ $C-\alpha_i-\beta_i=0$

As $\beta$ appears ony in constraints, we can write as: $0\leq\alpha_i\leq C$

## Answer

The equation for standard SVM with bias is given as:
$$ \min_{w}   \frac{1}{2}  ||w||^2 $$ 
subject to: $$y_i(w^Tx_i+b)\geq 1,   i=1,2,...N  $$
On comparing with soft-margin SVM, slack variables are missing in the above equations.

**Saddle point conditions:**
For standard SVM, saddle point conditions are as follows:
$$ w=\sum_{i=1}^{N}\alpha_i y_i x_i$$
$$C-\alpha_i-\beta_i=0$$
$$ \frac{\partial\Lambda}{\partial b}=\sum_{i=1}^{N}\alpha_iy_i=0$$
The above equation is not there in soft-margin SVM without bias.

**KKT conditions:**

For soft-margin SVM, $ \alpha_i [1 - \xi_i - y_i(w^Tx_i)] = 0 $ 

For standard SVM, $ \alpha_i [1 - y_i(w^Tx_i+b)] = 0 $

Since, $\alpha_i > 0$
$$y_i(w^Tx_i+b)=1$$
$$b=y_i-w^Tx_i$$
The above equation is not there in soft-margin SVM without bias.

**Dual Equations:**
The dual equation for standard SVM is similar to that of soft-margin SVM except that the constraints are different.
$$\max_{\alpha} \sum_{i=1}^{N}\alpha_i - \frac{1}{2}\sum_{i=1}^{N}\sum_{j=1}^{N}\alpha_i\alpha_jy_iy_jx_i^Tx_j$$
subject to $ \alpha_i\geq0,$ $\sum_{i=1}^{N}\alpha_iy_i=0$ 




## Part 2:  Soft-margin SVM for separable data 

Suppose you are given a linearly separable dataset, and you are training the soft-margin SVM, which uses slack variables with the soft-margin constant $C$ set 
to some positive value.
Consider the following statement:

Since increasing the $\xi_i$ can only increase the cost function of the primal problem (which
we are trying to minimize), at the solution to the primal problem, i.e. the hyperplane that minimizes the primal cost function, all the
training examples will have $\xi_i$ equal
to zero. 

Is this true or false?  Explain!

## Answer

The given statement is false.
In soft-margin case, SVM allows misclassification i.e. it gives some relaxation to few points. Thus, instead of considering those data points (that make our margin small and descision boundary poor) as support vectors, they are considered as error points and a certain penalty is given.
For soft-margin SVM, we are minimizing:
$$\min \frac{1}{2}||w||^2 + C\sum_{i=1}^{N}\xi_i$$ 
subject to: $ y_i(w^Tx_i+b)\geq1-\xi_i$; $\xi_i\geq0$; for $i=1,2,...,N$.

$\xi$ helps in allowing some slackness in constraint. 

For linearly separable data, there are possibilities that some data will reside in margin or can be classified correctly. Thus, $\xi$ can have non-zero values for some of the training examples.

## Import Libraries

````python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.datasets import load_svmlight_file
from sklearn import cross_validation
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_validate
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.linear_model import LogisticRegressionCV
import matplotlib.pyplot as plt
%matplotlib inline
````

## Read Dataset

````python
dataRead = pd.read_csv('scop_motif.data')
modifiedData = dataRead.iloc[:,1].str.split(" ", expand = True)
modifiedData[0] = np.where(modifiedData[0]=='rest', 1, -1)
filename_new = 'scop_motif_new.data'
modifiedData.to_csv(filename_new,sep=' ', index=False, header=False)
X, y = load_svmlight_file("scop_motif_new.data")
````

## Part 1

````python
##http://scikit-learn.org/stable/auto_examples/svm/plot_rbf_parameters.html
cv_gen = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
# Gaussian Kernel
C_param = [100, 10, 1, 0.1, 0.01]
Gamma_param = [0.0001, 0.001, 0.01, 0.1, 1]
grid_gaussian_parameter = dict(C=C_param,gamma=Gamma_param,kernel=['rbf'])
svm_gaussian_grid = GridSearchCV(estimator=SVC(), param_grid=grid_gaussian_parameter, cv=cv_gen, scoring='roc_auc') 
svm_gaussian_grid.fit(X,y)
#print(svm_gaussian_grid.cv_results_)
scores_gaussian = svm_gaussian_grid.cv_results_['mean_test_score'].reshape(len(C_param),len(Gamma_param))
print("Best parameters for Gaussian Kernel are %s with a score of %0.2f"% (svm_gaussian_grid.best_params_, svm_gaussian_grid.best_score_))

# Plot
plt.figure(figsize=(10, 8))
plt.subplots_adjust(left=.2, right=0.95, bottom=0.15, top=0.95)
plt.imshow(scores_gaussian, interpolation='nearest', cmap=plt.cm.gist_heat)
plt.xticks(np.arange(len(Gamma_param)), Gamma_param)
plt.yticks(np.arange(len(C_param)), C_param)
plt.title('Gaussian SVM')
plt.xlabel('Gamma')
plt.ylabel('Soft Margin parameter (C)')
plt.colorbar()

# Polynomial Kernel
Cpoly_param = [100, 10, 1, 0.1, 0.01]
degree_param = [1, 2, 3, 4, 5]
grid_poly_parameter = dict(C=Cpoly_param,degree=degree_param,kernel=['poly'])
svm_poly_grid = GridSearchCV(estimator=SVC(coef0=1), param_grid=grid_poly_parameter, cv=cv_gen, scoring='roc_auc') 
svm_poly_grid.fit(X,y)
#print(svm_poly_grid.cv_results_)
scores_poly = svm_poly_grid.cv_results_['mean_test_score'].reshape(len(Cpoly_param),len(degree_param))
print("Best parameters for Polynomial Kernel are %s with a score of %0.2f"% (svm_poly_grid.best_params_, svm_poly_grid.best_score_))

# Plot
plt.figure(figsize=(10, 8))
plt.subplots_adjust(left=.2, right=0.95, bottom=0.15, top=0.95)
plt.imshow(scores_poly, interpolation='nearest', cmap=plt.cm.gist_heat)
plt.xticks(np.arange(len(degree_param)), degree_param)
plt.yticks(np.arange(len(Cpoly_param)), Cpoly_param)
plt.title('Polynomial SVM')
plt.xlabel('Polynomial degree')
plt.ylabel('Soft Margin parameter (C)')
plt.colorbar()
plt.show()
````

## Plot for Cross-Sections

````python
# Gaussian Kernel
C_param = [0.01,0.1,1,10,100]
Gamma_param = [0.0001, 0.001, 0.01, 0.1, 1]

scores_gamma=scores_gaussian[:,0]
scores_gaussian1=scores_gamma[::-1]
plt.figure(figsize=(6, 6))
plt.subplot(2, 1, 1)
plt.semilogx(C_param,scores_gaussian1,basex=10,label='ROC curve')
plt.title('Gaussian Kernel with gamma=0.0001')
plt.xlabel('Soft Margin parameter (C)')
plt.ylabel('Accuracy (ROC AUC)')
plt.legend(loc='best')

plt.figure(figsize=(6, 6))
plt.subplot(2, 1, 2)
scores_C=scores_gaussian[1,]
plt.semilogx(Gamma_param,scores_C,basex=10,label='ROC curve')
plt.title('Gaussian Kernel with C=10')
plt.xlabel('Gamma')
plt.ylabel('Accuracy (ROC AUC)')
plt.legend(loc='best')
````

## Observation

For a given value of gamma, in case of gaussian kernel, accuracy increases with increasing value of C and then starts decreasing. Whereas, accuracy decreases with increase in gamma for a given C.

````python
# Polynomial Kernel
C_param = [0.01,0.1,1,10,100]
degree_param = [1, 2, 3, 4, 5]

plt.figure(figsize=(6, 6))
plt.subplot(2, 1, 1)
scores_degree=scores_poly[:,4]
scores_poly1=scores_degree[::-1]
plt.semilogx(C_param,scores_poly1,basex=10,label='ROC curve')
plt.title('Polynomial Kernel with degree=5')
plt.xlabel('Soft Margin parameter (C)')
plt.ylabel('Accuracy (ROC AUC)')
plt.legend(loc='best')

plt.figure(figsize=(6, 6))
plt.subplot(2, 1, 2)
scores_degree=scores_poly[0,]
plt.plot(degree_param,scores_degree,label='ROC curve')
plt.title('Polynomial Kernel with C=100')
plt.xlabel('Degree')
plt.ylabel('Accuracy (ROC AUC)')
plt.legend(loc='best')
````

## Observation

In case of polynomial kernel, accuracy increases with increase in both C and degree values.


## Accuracy Comparison

````python
## Normalizing Dataset
norm_data=preprocessing.normalize(X, norm='l2')
X_norm = norm_data

# nested cross validation code from 06_model_selection notebook
#Gaussian Kernel
param_grid_gaussian = [
  {'C': [0.1, 1, 10, 100], 'gamma': [0.1, 1, 10, 100], 'kernel': ['rbf']}
]
gaussian_classifier = GridSearchCV(SVC(), param_grid_gaussian, cv=5)
#Gaussian Kernel(Original data)
cv_results_nested2 = cross_validate(gaussian_classifier, X, y, cv=cv_gen, 
                                  scoring=('roc_auc'))
raw_gaussian=np.mean(cv_results_nested2['test_score'])
#Gaussian Kernel(Normalized data)
cv_results_nested2 = cross_validate(gaussian_classifier, X_norm, y, cv=cv_gen, 
                                  scoring=('roc_auc'))
norm_gaussian=np.mean(cv_results_nested2['test_score'])
#Linear Kernel
param_grid_linear = [
  {'C': [0.1, 1, 10, 100], 'kernel': ['linear']}
]
linear_classifier = GridSearchCV(SVC(), param_grid_linear, cv=5)
#Linear Kernel(Original data)
cv_results_nested2 = cross_validate(linear_classifier, X, y, cv=cv_gen, 
                                  scoring=('roc_auc'))
raw_linear=np.mean(cv_results_nested2['test_score'])
#Linear Kernel(Normalized data)
cv_results_nested2 = cross_validate(linear_classifier, X_norm, y, cv=cv_gen, 
                                  scoring=('roc_auc'))
norm_linear=np.mean(cv_results_nested2['test_score'])

# Regularized Logistic Regression
lr_classifier = LogisticRegressionCV(random_state=0, solver='lbfgs', multi_class='multinomial', Cs = [0.1, 1, 10, 100], cv=5)
# Regularized Logistic Regression(Original data)
cv_results_nested2 = cross_validate(lr_classifier, X, y, cv=cv_gen, 
                                  scoring=('roc_auc'))
raw_logistic=np.mean(cv_results_nested2['test_score'])
# Regularized Logistic regression(Normalized data)
cv_results_nested2 = cross_validate(lr_classifier, X_norm, y, cv=cv_gen, 
                                  scoring=('roc_auc'))
norm_logistic=np.mean(cv_results_nested2['test_score'])

#Display Comparison results
titles = ['Gaussian', 'Linear', 'Regularized Logistic Regression']
test_scores_original = pd.Series([raw_gaussian,
                            raw_linear,
                            raw_logistic],
                            index=titles)
test_scores_norm = pd.Series([norm_gaussian,
                            norm_linear,
                            norm_logistic],
                            index=titles)
d = {'Original Data' : test_scores_original, 'Normalized Data' : test_scores_norm}
dataframe = pd.DataFrame(d)
dataframe
````
